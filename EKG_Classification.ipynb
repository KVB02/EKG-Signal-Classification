```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
P_TR = pd.read_csv('P_TR.csv',header = None).values
P_TS = pd.read_csv('P_TS.csv',header = None).values
P_TT = pd.read_csv('P_TT.csv',header = None).values
T_TR = pd.read_csv('T_TR.csv',header = None).values.flatten()
T_TS = pd.read_csv('T_TS.csv',header = None).values.flatten()
T_TT = pd.read_csv('T_TT.csv',header = None).values.flatten()
```

```python
def tansig(x):
    return (2 / (1 + np.exp(-2*x))) - 1

def fdtansig(x):
    return 1 - (tansig(x)**2)
```

```python
# Set the number of input features, hidden units, and output units
num_input_units = 47
alpha_values = [0.001, 0.1]

# Define a list of different numbers of hidden units to iterate over
hidden_units_values = [100, 75]

# Maximum number of training epochs
max_epochs = 100

# Initialize dictionaries to track Mean Squared Error 
mse_alpha_hidden = {}

# Filter all warnings
warnings.filterwarnings("ignore")

for alpha in alpha_values:
    for num_hidden_units in hidden_units_values:
        # Initialize weights and bias to zero for each alpha and hidden unit value
        # Initialize weights and biases for the hidden layer
        W1 = np.random.randn(num_hidden_units, num_input_units)
        b1 = np.random.randn(num_hidden_units, 1)
        # Initialize weights and biases for the output layer
        W2 = np.random.randn(1, num_hidden_units)
        b2 = np.random.randn(1, 1)
        
        # Initialize dictionaries to track Mean Squared Error for this alpha and hidden unit value
        mse_values_train = []
        mse_values_val = []
        
        for epoch in range(max_epochs):
        # PHASE I - Training
            mse_train = 0  # Initialize training MSE for this epoch
            for i, p in enumerate(P_TR.T):
                # Forward propagation for training
                n1 = np.dot(W1, p).reshape(-1, 1) + b1
                a1 = tansig(n1)
                n2 = np.dot(W2, a1).reshape(-1, 1) + b2
                a2 = tansig(n2)
                error = T_TR[i] - a2
                s2 = (-2) * (fdtansig(a2) * error)
                s1 = fdtansig(a1) * np.outer(W2, s2)
                W2 = W2 - alpha * np.outer(s2, a1)
                b2 = b2 - alpha * s2.reshape(-1, 1)
                W1 = W1 - alpha * np.outer(s1, p)
                b1 = b1 - alpha * s1.reshape(-1, 1)
                mse_train = mse_train + (error ** 2)  # Accumulate training squared error
            
            # Calculate, store and print the average Training MSE for the epoch
            average_mse_train = mse_train / len(P_TR.T)
            mse_values_train.append(average_mse_train)
            np.set_printoptions(precision=30)
            print(f"Epoch {epoch + 1}/{max_epochs}, Training MSE: {average_mse_train}")  
        
            # PHASE II - Validation
            mse_val = 0  # Initialize validation MSE for this epoch
            for i, p_val in enumerate(P_TT.T):
                # Forward propagation for validation
                n1_val = np.dot(W1, p_val).reshape(-1, 1) + b1
                a1_val = tansig(n1_val)
                n2_val = np.dot(W2, a1_val).reshape(-1, 1) + b2
                a2_val = tansig(n2_val)
                error_val = T_TT[i] - a2_val
                mse_val = mse_val + (error_val ** 2)  # Accumulate validation squared error

            # Calculate and store the average validation MSE for the epoch
            average_mse_val = mse_val / len(P_TT.T)
            mse_values_val.append(average_mse_val)
            print(f"Epoch {epoch + 1}/{max_epochs}, Validation MSE: {average_mse_val}")
        
        # Store the results for the current alpha
        mse_alpha_hidden[(alpha, num_hidden_units)] = {'train': mse_values_train, 'val': mse_values_val}
                

# Print the final MSE values for each alpha and hidden unit value
for (alpha, num_hidden_units), mse_values in mse_alpha_hidden.items():
    print(f"Final Training MSE for Alpha={alpha}, Hidden Units={num_hidden_units}: {mse_values['train'][-1]}")
    print(f"Final Validation MSE for Alpha={alpha}, Hidden Units={num_hidden_units}: {mse_values['val'][-1]}")
    
    
# Iterate over alpha values
for alpha in alpha_values:
    for num_hidden_units in hidden_units_values:
        plt.figure(figsize=(8, 4))

        # Training MSE
        mse_train_array = np.array(mse_alpha_hidden[(alpha, num_hidden_units)]['train']).reshape(-1)
        plt.plot(range(len(mse_train_array)), mse_train_array, label=f'Training MSE (Alpha={alpha}, Hidden Units={num_hidden_units})')

        # Validation MSE
        mse_val_array = np.array(mse_alpha_hidden[(alpha, num_hidden_units)]['val']).reshape(-1)
        plt.plot(range(len(mse_val_array)), mse_val_array, label=f'Validation MSE (Alpha={alpha}, Hidden Units={num_hidden_units})')

        plt.xlabel("Epoch")
        plt.ylabel("MSE")
        plt.title(f"MSE vs. Epoch (Alpha = {alpha}, Hidden Units = {num_hidden_units})")
        plt.legend()
        plt.show()
```

```python
mse_alpha_hidden_final = {}
# Define the learning rate (alpha)
alpha = 0.001

# Maximum number of training epochs
hidden_units = 75

# Filter all warnings
warnings.filterwarnings("ignore")

# Initialize weights and bias
W1 = np.random.randn(hidden_units, num_input_units)
b1 = np.random.randn(hidden_units, 1)
W2 = np.random.randn(1, hidden_units)
b2 = np.random.randn(1, 1)

# Lists to store MSE values during training and validation
mse_values_train = []
mse_values_val = []
final_w1 = None
finalw2 = None
finalb1 = None
finalb2 = None

term_epochs = 20

# Backpropagation training algorithm
for epoch in range(term_epochs):
    # PHASE I - Training
    mse_train = 0  # Initialize training MSE for this epoch
    for i, p in enumerate(P_TR.T):
                # Forward propagation for training
                n1 = np.dot(W1, p).reshape(-1, 1) + b1
                a1 = tansig(n1)
                n2 = np.dot(W2, a1).reshape(-1, 1) + b2
                a2 = tansig(n2)
                error = T_TR[i] - a2
                s2 = (-2) * (fdtansig(a2) * error)
                s1 = fdtansig(a1) * np.outer(W2, s2)
                W2 = W2 - alpha * np.outer(s2, a1)
                b2 = b2 - alpha * s2.reshape(-1, 1)
                W1 = W1 - alpha * np.outer(s1, p)
                b1 = b1 - alpha * s1.reshape(-1, 1)
                final_w1 = W1
                finalw2 = W2
                finalb1 = b1
                finalb2 = b2
                mse_train = mse_train + (error ** 2)  # Accumulate training squared error
            
    # Calculate, store and print the average Training MSE for the epoch
    average_mse_train = mse_train / len(P_TR.T)
    mse_values_train.append(average_mse_train)
    np.set_printoptions(precision=30)
    print(f"Epoch {epoch + 1}, Training MSE: {average_mse_train}")  
        
            # PHASE II - Validation
    mse_val = 0  # Initialize validation MSE for this epoch
    for i, p_val in enumerate(P_TT.T):
                # Forward propagation for validation
            n1_val = np.dot(W1, p_val).reshape(-1, 1) + b1
            a1_val = tansig(n1_val)
            n2_val = np.dot(W2, a1_val).reshape(-1, 1) + b2
            a2_val = tansig(n2_val)
            error_val = T_TT[i] - a2_val
            mse_val = mse_val + (error_val ** 2)  # Accumulate validation squared error

            # Calculate and store the average validation MSE for the epoch
    average_mse_val = mse_val / len(P_TT.T)
    mse_values_val.append(average_mse_val)
    print(f"Epoch {epoch + 1}, Validation MSE: {average_mse_val}")
    
    # Store the results for the current alpha
    mse_alpha_hidden_final[(alpha, hidden_units)] = {'train': mse_values_train, 'val': mse_values_val}
    

# Print the final MSE values for the alpha value
print(f"Final Training MSE for Alpha {alpha}: {mse_alpha_hidden_final[(alpha, hidden_units)]['train'][-1]}")
print(f"Final Validation MSE for Alpha {alpha}: {mse_alpha_hidden_final[(alpha, hidden_units)]['val'][-1]}")
```

```python
# Evaluate the best model on the test set
mse_test = 0
correct = 0;
incorrect = 0;
for i, p_test in enumerate(P_TS.T):
    # Forward propagation for testing using the best model
    n1_test = np.dot(final_w1, p_test).reshape(-1, 1) + finalb1
    a1_test = tansig(n1_test)
    n2_test = np.dot(finalw2, a1_test).reshape(-1, 1) + finalb2
    a2_test = tansig(n2_test)
    if a2_test < 0:
        output = -1
    else:
        output = 1
    if (T_TS[i] == output):
        correct = correct + 1
    else:
        incorrect = incorrect + 1
    error_test = T_TS[i] - a2_test
    mse_test = mse_test + (error_test ** 2)  # Accumulate testing squared error

# Calculate and print the average testing MSE for the best model
average_mse_test = mse_test / len(P_TS.T)
print(f"Final Test MSE for the Best Model: {average_mse_test}")
print(f'Number of Correct predictions : {correct}')
print(f'Number of Incorrect predictions :{incorrect}')
Hit_Ratio = correct/len(T_TS.T)
print(f'Hit and Ration : {Hit_Ratio}')
accuracy = Hit_Ratio * 100
print(f"Accuracy: {accuracy:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

# # Loading the datasets
# P_TR = pd.read_csv('/mnt/data/P_TR.csv', header=None).values.T
# P_TS = pd.read_csv('/mnt/data/P_TS.csv', header=None).values.T
# P_TT = pd.read_csv('/mnt/data/P_TT.csv', header=None).values.T
# T_TR = pd.read_csv('/mnt/data/T_TR.csv', header=None).values.flatten()
# T_TS = pd.read_csv('/mnt/data/T_TS.csv', header=None).values.flatten()
# T_TT = pd.read_csv('/mnt/data/T_TT.csv', header=None).values.flatten()

# Activation functions
def tansig(x):
    return (2 / (1 + np.exp(-2 * x))) - 1

def fdtansig(x):
    return 1 - tansig(x) ** 2

# Network configuration
num_features = 47
learning_rates = [0.01, 0.005]
momentum_values = [0.9,0.7]  # Momentum values to try
hidden_units_list = [100, 75]  # Specified hidden units
epochs_max = 120

# Initialize dictionaries to track Mean Squared Error and weights
mse_alpha_momentum = {}

# Suppress all warnings
warnings.filterwarnings("ignore")

# Backpropagation with Momentum training algorithm
for learning_rate in learning_rates:
    for momentum in momentum_values:
        for hidden_units in hidden_units_list:
            # Initialize weights and biases
            W1 = np.random.randn(hidden_units, num_features)
            b1 = np.random.randn(hidden_units, 1)
            W2 = np.random.randn(1, hidden_units)
            b2 = np.random.randn(1, 1)

            # Initialize previous weight updates for momentum term
            W1_update = np.zeros(W1.shape)
            b1_update = np.zeros(b1.shape)
            W2_update = np.zeros(W2.shape)
            b2_update = np.zeros(b2.shape)

            mse_values_train = []
            mse_values_val = []

            for epoch in range(max_epochs):
                # Training phase
                mse_train = 0
                for i, p in enumerate(P_TR.T):
                    # Forward propagation
                    n1 = np.dot(W1, p).reshape(-1, 1) + b1
                    a1 = tansig(n1)
                    n2 = np.dot(W2, a1).reshape(-1, 1) + b2
                    a2 = tansig(n2)
                    error = T_TR[i] - a2

                    # Backpropagation with Momentum
                    s2 = (-2) * (fdtansig(a2) * error)
                    s1 = fdtansig(a1) * np.dot(W2.T, s2)

                    W2_update = (momentum * W2_update) - (learning_rate * np.outer(s2, a1))
                    b2_update = (momentum * b2_update) - (learning_rate * s2)
                    W1_update = (momentum * W1_update) - (learning_rate * np.outer(s1, p))
                    b1_update = (momentum * b1_update) - (learning_rate * s1)

                    W2 += W2_update
                    b2 += b2_update.reshape(-1, 1)
                    W1 += W1_update
                    b1 += b1_update.reshape(-1, 1)

                    mse_train += np.sum(error ** 2)

                average_mse_train = mse_train / len(P_TR.T)
                mse_values_train.append(average_mse_train)

                # Validation phase
                mse_val = 0
                for i, p_val in enumerate(P_TT.T):
                    # Forward propagation
                    n1_val = np.dot(W1, p_val).reshape(-1, 1) + b1
                    a1_val = tansig(n1_val)
                    n2_val = np.dot(W2, a1_val).reshape(-1, 1) + b2
                    a2_val = tansig(n2_val)
                    error_val = T_TT[i] - a2_val
                    mse_val += np.sum(error_val ** 2)

                average_mse_val = mse_val / len(P_TT.T)
                mse_values_val.append(average_mse_val)

            # Store the results
            mse_alpha_momentum[(learning_rate, momentum, hidden_units)] = {'train': mse_values_train, 'val': mse_values_val}

# Plotting
for (learning_rate, momentum, hidden_units), mse_values in mse_alpha_momentum.items():
    plt.figure(figsize=(10, 5))
    plt.plot(mse_values['train'], label='Training MSE')
    plt.plot(mse_values['val'], label='Validation MSE')
    plt.title(f'Learning Rate: {learning_rate}, Momentum: {momentum}, Hidden Units: {hidden_units}')
    plt.xlabel('Epochs')
    plt.ylabel('MSE')
    plt.legend()
    plt.show()
```

```python
# 1. Selecting Optimal Hyperparameters
best_alpha, best_gamma, best_hidden_units = None, None, None
lowest_val_mse = float('inf')

term_epoch = 25

for key, mse_dict in mse_alpha_momentum.items():
    current_val_mse = mse_dict['val'][-1]  # Get the final validation MSE
    if current_val_mse < lowest_val_mse:
        lowest_val_mse = current_val_mse
        best_alpha, best_gamma, best_hidden_units = key

print(f"Optimal Hyperparameters - Alpha: {best_alpha}, Gamma: {best_gamma}, Hidden Units: {best_hidden_units}")

# 2. Retraining the Network with Optimal Hyperparameters
W1 = np.random.randn(best_hidden_units, num_features)
b1 = np.random.randn(best_hidden_units, 1)
W2 = np.random.randn(1, best_hidden_units)
b2 = np.random.randn(1, 1)

# Initialize previous weight updates for momentum term
W1_update = np.zeros(W1.shape)
b1_update = np.zeros(b1.shape)
W2_update = np.zeros(W2.shape)
b2_update = np.zeros(b2.shape)

mse_values_train_best = []
mse_values_val_best = []

for epoch in range(term_epoch):
    # Training phase
    mse_train = 0
    for i, p in enumerate(P_TR.T):
        # Forward propagation
        n1 = np.dot(W1, p).reshape(-1, 1) + b1
        a1 = tansig(n1)
        n2 = np.dot(W2, a1).reshape(-1, 1) + b2
        a2 = tansig(n2)
        error = T_TR[i] - a2

        # Backpropagation with Momentum
        s2 = (-2) * (fdtansig(a2) * error)
        s1 = fdtansig(a1) * np.dot(W2.T, s2)

        W2_update = (momentum * W2_update) - (learning_rate * np.outer(s2, a1))
        b2_update = (momentum * b2_update) - (learning_rate * s2)
        W1_update = (momentum * W1_update) - (learning_rate * np.outer(s1, p))
        b1_update = (momentum * b1_update) - (learning_rate * s1)

        W2 += W2_update
        b2 += b2_update.reshape(-1, 1)
        W1 += W1_update
        b1 += b1_update.reshape(-1, 1)

        mse_train += np.sum(error ** 2)

    average_mse_train = mse_train / len(P_TR.T)
    mse_values_train.append(average_mse_train)

    # Validation phase
    mse_val = 0
    for i, p_val in enumerate(P_TT.T):
        # Forward propagation
        n1_val = np.dot(W1, p_val).reshape(-1, 1) + b1
        a1_val = tansig(n1_val)
        n2_val = np.dot(W2, a1_val).reshape(-1, 1) + b2
        a2_val = tansig(n2_val)
        error_val = T_TT[i] - a2_val
        mse_val += np.sum(error_val ** 2)

        average_mse_val = mse_val / len(P_TT.T)
        mse_values_val.append(average_mse_val)

    # Store MSE for the epoch
    mse_values_train_best.append(mse_train / len(P_TR.T))
    mse_values_val_best.append(mse_val / len(P_TT.T))

# 3. Plotting Training and Validation MSE for the Best Hyperparameters
plt.figure(figsize=(10, 5))
plt.plot(mse_values_train_best, label='Training MSE')
plt.plot(mse_values_val_best, label='Validation MSE')
plt.title(f'Training and Validation MSE - Best Hyperparameters')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()

# Testing and Computing Hit Ratio
mse_test = 0
correct = 0
incorrect = 0

for i, p_test in enumerate(P_TS.T):
    # Forward propagation for testing using the best model
    n1_test = np.dot(W1, p_test).reshape(-1, 1) + b1
    a1_test = tansig(n1_test)
    n2_test = np.dot(W2, a1_test).reshape(-1, 1) + b2
    a2_test = tansig(n2_test)

    # Classify and compare with actual target
    predicted = -1 if a2_test < 0 else 1
    correct += (predicted == T_TS[i])
    incorrect += (predicted != T_TS[i])
    error_test = T_TS[i] - a2_test
    mse_test += (error_test ** 2)

# Calculate and print the average testing MSE and Hit Ratio
average_mse_test = mse_test / len(P_TS.T)
hit_ratio = correct / len(P_TS.T)
accuracy = hit_ratio * 100

print(f"Final Test MSE: {average_mse_test}")
print(f"Hit Ratio (Accuracy): {accuracy:.2f}%")
```
